````python
from pymongo import MongoClient
from datetime import datetime
from typing import List, Dict, Optional
import os

class HotelDatabase:
    def __init__(self, connection_string: str = None):
        if connection_string is None:
            connection_string = os.getenv("MONGODB_URI")

        print("üîó Connecting to MongoDB Atlas...")
        self.client = MongoClient(connection_string)
        self.db = self.client.hotel_service

        self.menu = self.db.menu
        self.orders = self.db.orders
        self.conversations = self.db.conversations
        self.guests = self.db.guests

        self._initialize_data()

    def _initialize_data(self):
        if self.menu.count_documents({}) == 0:
            menu_items = [
                {
                    "item_id": "BF001",
                    "name": "Masala Dosa",
                    "category": "breakfast",
                    "description": "Crispy rice crepe with potato masala and coconut chutney",
                    "price": 180,
                    "prep_time": 15,
                    "dietary_tags": ["vegetarian", "gluten-free"],
                    "allergens": [],
                    "available": True,
                    "stock": 50
                },
                {
                    "item_id": "BF002",
                    "name": "Poha",
                    "category": "breakfast",
                    "description": "Flattened rice with peanuts, onions and spices",
                    "price": 120,
                    "prep_time": 10,
                    "dietary_tags": ["vegetarian", "gluten-free"],
                    "allergens": ["peanuts"],
                    "available": True,
                    "stock": 30
                },
                {
                    "item_id": "BF003",
                    "name": "Aloo Paratha",
                    "category": "breakfast",
                    "description": "Whole wheat bread stuffed with spiced potatoes",
                    "price": 150,
                    "prep_time": 12,
                    "dietary_tags": ["vegetarian"],
                    "allergens": ["gluten", "dairy"],
                    "available": True,
                    "stock": 40
                },
                {
                    "item_id": "LN001",
                    "name": "Butter Chicken",
                    "category": "lunch",
                    "description": "Tender chicken in rich tomato butter gravy",
                    "price": 450,
                    "prep_time": 25,
                    "dietary_tags": ["non-vegetarian"],
                    "allergens": ["dairy"],
                    "available": True,
                    "stock": 20
                },
                {
                    "item_id": "LN002",
                    "name": "Palak Paneer",
                    "category": "lunch",
                    "description": "Cottage cheese in creamy spinach gravy",
                    "price": 320,
                    "prep_time": 20,
                    "dietary_tags": ["vegetarian"],
                    "allergens": ["dairy"],
                    "available": True,
                    "stock": 25
                },
                {
                    "item_id": "LN003",
                    "name": "Vegetable Biryani",
                    "category": "lunch",
                    "description": "Fragrant basmati rice with mixed vegetables and spices",
                    "price": 280,
                    "prep_time": 22,
                    "dietary_tags": ["vegetarian"],
                    "allergens": [],
                    "available": True,
                    "stock": 35
                },
                {
                    "item_id": "DN001",
                    "name": "Dal Tadka",
                    "category": "dinner",
                    "description": "Lentils tempered with garlic and spices",
                    "price": 190,
                    "prep_time": 18,
                    "dietary_tags": ["vegetarian", "gluten-free", "vegan"],
                    "allergens": [],
                    "available": True,
                    "stock": 40
                },
                {
                    "item_id": "DN002",
                    "name": "Chicken Curry",
                    "category": "dinner",
                    "description": "Chicken in traditional Indian gravy with spices",
                    "price": 380,
                    "prep_time": 28,
                    "dietary_tags": ["non-vegetarian"],
                    "allergens": [],
                    "available": True,
                    "stock": 18
                },
                {
                    "item_id": "BV001",
                    "name": "Masala Chai",
                    "category": "beverages",
                    "description": "Traditional Indian spiced tea with milk",
                    "price": 60,
                    "prep_time": 5,
                    "dietary_tags": ["vegetarian"],
                    "allergens": ["dairy"],
                    "available": True,
                    "stock": 100
                },
                {
                    "item_id": "BV002",
                    "name": "Mango Lassi",
                    "category": "beverages",
                    "description": "Sweet yogurt drink with mango pulp",
                    "price": 120,
                    "prep_time": 3,
                    "dietary_tags": ["vegetarian"],
                    "allergens": ["dairy"],
                    "available": True,
                    "stock": 50
                },
                {
                    "item_id": "BV003",
                    "name": "Mineral Water",
                    "category": "beverages",
                    "description": "1 liter bottled mineral water",
                    "price": 40,
                    "prep_time": 1,
                    "dietary_tags": ["vegan", "gluten-free", "dairy-free"],
                    "allergens": [],
                    "available": True,
                    "stock": 200
                }
            ]
            self.menu.insert_many(menu_items)
            print("‚úÖ Menu initialized successfully!")
        else:
            print("‚úÖ Database already initialized!")

    def get_menu_items(self, category: str = None, available: bool = True) -> List[Dict]:
        query = {"available": available}
        if category:
            query["category"] = category
        return list(self.menu.find(query, {"_id": 0}))

    def find_menu_item(self, name: str) -> Optional[Dict]:
        return self.menu.find_one({
            "name": {"$regex": f"^{name}$", "$options": "i"},
            "available": True
        }, {"_id": 0})

    def search_menu_items(self, search_term: str) -> List[Dict]:
        return list(self.menu.find({
            "$or": [
                {"name": {"$regex": search_term, "$options": "i"}},
                {"description": {"$regex": search_term, "$options": "i"}},
                {"dietary_tags": {"$in": [search_term.lower()]}}
            ],
            "available": True
        }, {"_id": 0}))

    def create_order(self, order_data: Dict) -> str:
        result = self.orders.insert_one(order_data)
        return str(result.inserted_id)

    def update_item_stock(self, item_id: str, quantity: int):
        self.menu.update_one(
            {"item_id": item_id},
            {"$inc": {"stock": -quantity}}
        )

    def save_conversation(self, guest_id: str, role: str, message: str, metadata: Dict = None):
        conversation_data = {
            "guest_id": guest_id,
            "role": role,
            "message": message,
            "timestamp": datetime.now(),
            "metadata": metadata or {}
        }
        self.conversations.insert_one(conversation_data)

    def get_conversation_history(self, guest_id: str, limit: int = 10) -> List[Dict]:
        return list(self.conversations.find(
            {"guest_id": guest_id},
            {"_id": 0}
        ).sort("timestamp", -1).limit(limit))

    def update_guest_profile(self, guest_id: str, preferences: List[str], allergies: List[str]):
        self.guests.update_one(
            {"guest_id": guest_id},
            {
                "$set": {
                    "dietary_preferences": preferences,
                    "allergies": allergies,
                    "last_updated": datetime.now()
                }
            },
            upsert=True
        )

    def get_guest_profile(self, guest_id: str) -> Optional[Dict]:
        return self.guests.find_one({"guest_id": guest_id}, {"_id": 0})

# Test the connection
try:
    db = HotelDatabase()
    print("‚úÖ MongoDB Atlas connection successful!")
    print(f"üìä Menu items: {db.menu.count_documents({})}")
    print(f"üìä Orders: {db.orders.count_documents({})}")
    print(f"üìä Conversations: {db.conversations.count_documents({})}")
    print(f"üìä Guests: {db.guests.count_documents({})}")
except Exception as e:
    print(f"‚ùå MongoDB connection failed: {e}")
````

````python
import google.generativeai as genai
import os
import json
from typing import Dict, List, Optional, Tuple
from datetime import datetime

class RoomServiceAgent:
    def __init__(self):
        api_key = os.getenv("GEMINI_API_KEY")
        if not api_key:
            raise ValueError("GEMINI_API_KEY environment variable is required")

        genai.configure(api_key=api_key)
        self.model = genai.GenerativeModel('gemini-pro')
        self.db = HotelDatabase()

        self.system_prompt = """
        You are a professional hotel room service assistant at a luxury hotel in India.
        Your role is to help guests place food and beverage orders with excellent hospitality.

        CORE RESPONSIBILITIES:
        1. Process natural language orders and understand guest intent
        2. Handle dietary preferences (vegetarian, vegan, gluten-free, Jain, etc.)
        3. Manage allergies and dietary restrictions carefully
        4. Ask clarifying questions when orders are ambiguous
        5. Provide accurate menu information and recommendations
        6. Confirm orders before finalizing
        7. Maintain friendly, professional tone

        IMPORTANT RULES:
        - Always check for dietary conflicts before confirming orders
        - If unsure about anything, ask clarifying questions
        - Provide estimated preparation times
        - Be culturally sensitive to Indian dietary preferences
        - Handle multi-turn conversations naturally

        Respond in a helpful, warm, and professional manner.
        """

    def get_menu_context(self) -> str:
        menu_items = self.db.get_menu_items()
        menu_text = "CURRENT MENU:\n\n"

        categories = {}
        for item in menu_items:
            cat = item["category"]
            if cat not in categories:
                categories[cat] = []
            categories[cat].append(item)

        for category, items in categories.items():
            menu_text += f"=== {category.upper()} ===\n"
            for item in items:
                dietary = ", ".join(item["dietary_tags"]) if item["dietary_tags"] else "Regular"
                allergens = ", ".join(item["allergens"]) if item["allergens"] else "None"
                menu_text += f"‚Ä¢ {item['name']} (‚Çπ{item['price']})\n"
                menu_text += f"  Description: {item['description']}\n"
                menu_text += f"  Dietary: {dietary} | Allergens: {allergens}\n"
                menu_text += f"  Prep Time: {item['prep_time']} mins\n\n"

        return menu_text

    def process_guest_message(self, guest_id: str, message: str, room_number: str = None, guest_name: str = None) -> Dict:
        # Get conversation history
        history = self.db.get_conversation_history(guest_id)
        history_text = "CONVERSATION HISTORY:\n"
        for conv in reversed(history[-6:]):  # Last 6 messages
            history_text += f"{conv['role'].title()}: {conv['message']}\n"

        # Get guest profile
        guest_profile = self.db.get_guest_profile(guest_id) or {}
        dietary_prefs = guest_profile.get('dietary_preferences', [])
        allergies = guest_profile.get('allergies', [])

        menu_context = self.get_menu_context()

        prompt = f"""
        {self.system_prompt}

        {menu_context}

        {history_text}

        GUEST PROFILE:
        - Dietary Preferences: {', '.join(dietary_prefs) if dietary_prefs else 'Not specified'}
        - Allergies: {', '.join(allergies) if allergies else 'None reported'}
        - Room: {room_number or 'Not specified'}
        - Name: {guest_name or 'Not specified'}

        Current guest message: "{message}"

        Analyze this message and:
        1. Identify ordered items and quantities
        2. Detect any dietary preferences or allergies mentioned
        3. Determine if clarification is needed
        4. Check for stock availability and dietary conflicts
        5. Provide appropriate response

        Respond in JSON format:
        {{
            "interpretation": {{
                "ordered_items": [
                    {{
                        "name": "item name",
                        "quantity": 1,
                        "special_instructions": "any modifications"
                    }}
                ],
                "detected_preferences": ["list", "of", "preferences"],
                "detected_allergies": ["list", "of", "allergies"],
                "needs_clarification": true/false,
                "clarification_questions": ["question1", "question2"],
                "dietary_conflicts": ["conflict1", "conflict2"],
                "stock_issues": ["issue1", "issue2"]
            }},
            "response": {{
                "message": "your response to guest",
                "suggested_actions": ["action1", "action2"],
                "estimated_prep_time": 0,
                "total_price": 0
            }},
            "next_step": "continue|confirm|complete"
        }}
        """

        try:
            response = self.model.generate_content(prompt)
            response_text = response.text.strip()

            # Extract JSON from response
            if "```json" in response_text:
                json_str = response_text.split("```json")[1].split("```")[0]
            else:
                json_str = response_text

            result = json.loads(json_str)

            # Save conversation to database
            self.db.save_conversation(guest_id, "guest", message, {
                "interpretation": result.get("interpretation", {})
            })
            self.db.save_conversation(guest_id, "assistant", result["response"]["message"], {
                "analysis": result
            })

            # Update guest profile if new preferences/allergies detected
            new_prefs = result["interpretation"].get("detected_preferences", [])
            new_allergies = result["interpretation"].get("detected_allergies", [])
            if new_prefs or new_allergies:
                all_prefs = list(set(dietary_prefs + new_prefs))
                all_allergies = list(set(allergies + new_allergies))
                self.db.update_guest_profile(guest_id, all_prefs, all_allergies)

            return result

        except Exception as e:
            error_response = {
                "interpretation": {
                    "ordered_items": [],
                    "needs_clarification": True,
                    "clarification_questions": ["I apologize for the technical difficulty. Could you please repeat your order?"]
                },
                "response": {
                    "message": "I'm experiencing some technical difficulties. Could you please repeat your order?",
                    "suggested_actions": ["retry"]
                },
                "next_step": "continue"
            }
            self.db.save_conversation(guest_id, "assistant", error_response["response"]["message"])
            return error_response

    def confirm_order(self, guest_id: str, room_number: str, guest_name: str, final_items: List[Dict]) -> Dict:
        """Finalize and save the order"""

        # Calculate total and prep time
        total_price = 0
        max_prep_time = 0
        order_details = []

        for item in final_items:
            menu_item = self.db.find_menu_item(item["name"])
            if menu_item:
                quantity = item.get("quantity", 1)
                total_price += menu_item["price"] * quantity
                max_prep_time = max(max_prep_time, menu_item["prep_time"])
                order_details.append({
                    "item_id": menu_item["item_id"],
                    "name": menu_item["name"],
                    "quantity": quantity,
                    "price": menu_item["price"],
                    "special_instructions": item.get("special_instructions", "")
                })

        # Create order document
        order_data = {
            "order_id": f"ORD{datetime.now().strftime('%Y%m%d%H%M%S')}",
            "guest_id": guest_id,
            "guest_name": guest_name,
            "room_number": room_number,
            "items": order_details,
            "total_price": total_price,
            "estimated_prep_time": max_prep_time,
            "status": "confirmed",
            "order_time": datetime.now(),
            "delivery_notes": f"Deliver to Room {room_number}"
        }

        # Save order
        order_id = self.db.create_order(order_data)

        # Update stock
        for item in order_details:
            self.db.update_item_stock(item["item_id"], item["quantity"])

        confirmation_message = f"""
 ORDER CONFIRMED!

Thank you, {guest_name}!
Order ID: {order_id}
Room: {room_number}

Your order:
{chr(10).join([f"‚Ä¢ {item['name']} x{item['quantity']} - ‚Çπ{item['price'] * item['quantity']}" for item in order_details])}

Total: ‚Çπ{total_price}
Estimated Preparation Time: {max_prep_time} minutes

Your food will be delivered to your room shortly.
Enjoy your meal! üçõ
"""

        self.db.save_conversation(guest_id, "system", f"Order confirmed: {order_id}", {
            "order_details": order_data
        })

        return {
            "success": True,
            "order_id": order_id,
            "message": confirmation_message,
            "order_details": order_data
        }

print("AI Agent initialized successfully!")
````

````python
class HotelServiceSystem:
    def __init__(self):
        self.agent = RoomServiceAgent()
        self.db = HotelDatabase()
        self.active_sessions = {}

    def start_guest_session(self, guest_id: str, room_number: str, guest_name: str):
        self.active_sessions[guest_id] = {
            "room_number": room_number,
            "guest_name": guest_name,
            "current_order": [],
            "conversation_state": "active"
        }

        welcome_message = f"""
Namaste {guest_name}! üôè

Welcome to Hotel Grand Palace room service.
I'm your digital assistant, here to help you order delicious food and beverages.

How may I assist you today?
"""

        self.db.save_conversation(guest_id, "system", f"Session started for {guest_name}, Room {room_number}")
        self.db.save_conversation(guest_id, "assistant", welcome_message)

        return welcome_message

    def process_interaction(self, guest_id: str, user_input: str) -> str:
        if guest_id not in self.active_sessions:
            return "Please start a session first using start_guest_session()"

        session = self.active_sessions[guest_id]

        # Process with AI agent
        result = self.agent.process_guest_message(
            guest_id,
            user_input,
            session["room_number"],
            session["guest_name"]
        )

        response = result["response"]["message"]

        # Update current order if items are identified
        ordered_items = result["interpretation"].get("ordered_items", [])
        if ordered_items:
            session["current_order"].extend(ordered_items)

        # Show order summary if we have items
        if session["current_order"]:
            item_count = len(session["current_order"])
            total = sum(
                item.get("quantity", 1) * self._get_item_price(item["name"])
                for item in session["current_order"]
            )
            response += f"\n\nüì¶ Current Order: {item_count} item(s) - ‚Çπ{total}"

        return response

    def _get_item_price(self, item_name: str) -> int:
        item = self.db.find_menu_item(item_name)
        return item["price"] if item else 0

    def complete_order(self, guest_id: str) -> str:
        if guest_id not in self.active_sessions:
            return "No active session found."

        session = self.active_sessions[guest_id]

        if not session["current_order"]:
            return "Your order is empty. Please add some items first."

        result = self.agent.confirm_order(
            guest_id,
            session["room_number"],
            session["guest_name"],
            session["current_order"]
        )

        if result["success"]:
            # Clear current order
            session["current_order"] = []
            session["conversation_state"] = "completed"
            return result["message"]
        else:
            return "There was an issue confirming your order. Please try again."

    def show_menu(self, category: str = None) -> str:
        items = self.db.get_menu_items(category)

        if not items:
            return "No items found in this category."

        menu_text = "üçΩÔ∏è  OUR MENU  üçΩÔ∏è\n\n"

        categories = {}
        for item in items:
            cat = item["category"]
            if cat not in categories:
                categories[cat] = []
            categories[cat].append(item)

        for cat, cat_items in categories.items():
            menu_text += f"üìã {cat.upper()}\n"
            for item in cat_items:
                dietary_tags = " | ".join(item["dietary_tags"])
                menu_text += f"‚Ä¢ {item['name']} - ‚Çπ{item['price']}\n"
                menu_text += f"  {item['description']}\n"
                menu_text += f"  üè∑Ô∏è {dietary_tags} | ‚è±Ô∏è {item['prep_time']}min\n"
                if item["allergens"]:
                    menu_text += f"  ‚ö†Ô∏è Contains: {', '.join(item['allergens'])}\n"
                menu_text += "\n"

        return menu_text

# Test the 4 conversation flows
def test_conversation_flows():
    """Test 4 different conversation flows as required"""

    agent = RoomServiceAgent()
    db = HotelDatabase()

    test_cases = [
        {
            "name": "Flow 1: Simple Breakfast Order with Clarification",
            "guest_id": "test_guest_1",
            "room": "301",
            "guest_name": "Raj Sharma",
            "messages": [
                "I want breakfast",
                "Masala dosa and chai",
                "Yes, that's all"
            ]
        },
        {
            "name": "Flow 2: Dietary Restrictions & Allergy Management",
            "guest_id": "test_guest_2",
            "room": "415",
            "guest_name": "Priya Patel",
            "messages": [
                "I'm vegetarian and allergic to dairy",
                "What lunch options do you have?",
                "I'll have vegetable biryani and a mango lassi",
                "No, make the lassi without yogurt"
            ]
        },
        {
            "name": "Flow 3: Complex Family Order with Modifications",
            "guest_id": "test_guest_3",
            "room": "212",
            "guest_name": "The Kumar Family",
            "messages": [
                "We want butter chicken, palak paneer, and two naans",
                "Also one dal tadka without garlic",
                "And three mango lassis, make one sugar-free",
                "That's all for now"
            ]
        },
        {
            "name": "Flow 4: Ambiguous Request & Menu Exploration",
            "guest_id": "test_guest_4",
            "room": "508",
            "guest_name": "Mr. Gupta",
            "messages": [
                "I want something light for dinner",
                "What's in dal tadka?",
                "Is it spicy?",
                "Okay, I'll have that with one roti"
            ]
        }
    ]

    for i, test_case in enumerate(test_cases, 1):
        print(f"\n{'='*80}")
        print(f"CONVERSATION FLOW {i}: {test_case['name']}")
        print(f"{'='*80}")

        # Clear previous test data
        db.conversations.delete_many({"guest_id": test_case["guest_id"]})
        db.guests.delete_many({"guest_id": test_case["guest_id"]})

        print(f"Guest: {test_case['guest_name']} (Room {test_case['room']})")
        print("-" * 40)

        for j, message in enumerate(test_case["messages"], 1):
            print(f"\nTurn {j}:")
            print(f"Guest: {message}")

            result = agent.process_guest_message(
                test_case["guest_id"],
                message,
                test_case["room"],
                test_case["guest_name"]
            )

            print(f"Agent: {result['response']['message']}")

            # Show analysis
            interpretation = result.get('interpretation', {})
            if interpretation.get('needs_clarification'):
                print(f"üìù Clarification needed: {interpretation.get('clarification_questions', [])}")
            if interpretation.get('dietary_conflicts'):
                print(f"‚ö†Ô∏è Dietary conflicts: {interpretation.get('dietary_conflicts')}")
            if interpretation.get('ordered_items'):
                print(f"üõí Items detected: {[item['name'] for item in interpretation['ordered_items']]}")

        print(f"\nüéØ Conversation completed for {test_case['name']}")
        print(f"üíæ Saved to MongoDB Atlas with guest_id: {test_case['guest_id']}")

# Run the test flows
print("üöÄ Running 4 Conversation Flow Tests on MongoDB Atlas...")
test_conversation_flows()
````
